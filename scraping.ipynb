## 1. Import Library
Library utama yang dipakai adalah apify_client untuk scraping dan pandas untuk olah data.

```python
from apify_client import ApifyClient
import pandas as pd
```

## 2. Konfigurasi Token API dan URL Target

Isi TOKEN_API dengan API token Apify Anda, lalu masukkan URL video TikTok pada list VIDEO_URLS.

```python

TOKEN_API = "                   "
    
VIDEO_URLS = [
    "               "
]

client = ApifyClient(TOKEN_API)
```

## 3. Menyusun Parameter Scraping

```python
# Prepare input dengan multiple URLs
run_input = {
    "commentsPerPost": 700,
    "excludePinnedPosts": False,
    "maxRepliesPerComment": 7,
    "postURLs": VIDEO_URLS,
    "resultsPerPage": 100
}
```

## 4. Menjalankan Proses Scraping di Apify

```python
run = client.actor("BDec00yAmCm1QbMEI").call(run_input=run_input)
```

## 5. Mengambil Hasil Dataset dan Menyimpan ke CSV

```python

data_list = []

for item in client.dataset(run["defaultDatasetId"]).iterate_items():
    data_list.append({
        'comment_id': item.get('cid', ''),
        'username': item.get('uniqueId', ''),
        'user_id': item.get('uid', ''),
        'comment_text': item.get('text', ''),
        'likes': item.get('diggCount', 0),
        'create_time': item.get('createTimeISO', ''),
        'reply_count': item.get('replyCommentTotal', 0),
        'is_reply': item.get('repliesToId') is not None,
    })

df = pd.DataFrame(data_list)

main_comments = df[df['is_reply'] == False]
replies = df[df['is_reply'] == True]

df.to_csv("Tiktok.csv", index=False, encoding='utf-8')
```
